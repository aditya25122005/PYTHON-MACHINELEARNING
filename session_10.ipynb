```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri_data=sns.load_dataset('fmri')
fmri_data
```

```python

plt.figure(figsize=(10,4))
sns.lineplot(fmri_data,x='timepoint',y='signal')
```

```python

plt.figure(figsize=(10,4))
sns.lineplot(fmri_data,x='timepoint',y='signal',hue='subject')
```

```python

plt.figure(figsize=(10,4))
sns.lineplot(fmri_data,x='timepoint',y='signal',hue='event',style='region',markers=True)
```

```python
tips_data=sns.load_dataset('tips')
sns.jointplot(tips_data,x='total_bill',y='tip')
```

```python
sns.jointplot(tips_data,x='total_bill',y='tip',kind='hex')
```

```python
sns.jointplot(tips_data,x='total_bill',y='tip',kind='scatter',hue='sex')
```

```python
sns.pairplot(tips_data)
```

```python
sns.pairplot(tips_data,hue='time',palette='viridis')
```

```python
sns.pairplot(tips_data[['total_bill','tip','time']],hue='time')
```

```python
sns.pairplot(tips_data,hue='time',palette='viridis',x_vars=['total_bill','tip'],y_vars=['size'])
```

```python
sns.boxplot(tips_data,x='day',y='total_bill',hue='sex')
#Horizontal line is median
#median
#sort the data --Middle value (n+1)/2 th term

#Outlier-->abnormal value out of range
```

```python
sns.violinplot(tips_data,x='day',y='total_bill')
plt.grid()
```

```python
sns.stripplot(tips_data,x='day',y='total_bill')
```

```python
sns.swarmplot(tips_data,x='day',y='total_bill',color='g')
```
