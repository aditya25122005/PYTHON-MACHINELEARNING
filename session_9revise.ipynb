```python
# figsize--> dimensions of plot(12,6) 12 is width and 6 is height
import matplotlib.pyplot as plt
X=[1,2,3,4,5]
Y=[50,78,84,120,135]
fig,axes=plt.subplots(2,2,figsize=(12,6))
axes[0,0].plot(X,Y)
axes[0,0].grid()
axes[0,0].set_title("wave")

axes[0,1].plot(Y,X)
```

```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([10,20,30])
Y=np.array([80,60,100])
plt.bar(X,Y,width=3,color='r',tick_label=['oil','rice','sugar'])
plt.title('Goods Prices')
plt.ylabel('Prices')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([10,20,30])
Y=np.array([80,60,100])
Y2025=np.array([50,140,80])
plt.bar(X+1.5,Y,width=3,color='r',tick_label=['oil','rice','sugar'],label='2024')
plt.bar(X-1.5,Y2025,width=3,color='b',tick_label=['oil','rice','sugar'],label='2025')
plt.title('Goods Prices')
plt.ylabel('Prices')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([10,20,30])
Y=np.array([80,60,100])
Y2025=np.array([50,140,80])
plt.bar(X+1.5,Y,width=3,color='r',label='2024')
plt.bar(X-1.5,Y2025,width=3,color='b',label='2025')
plt.title('Goods Prices')
plt.xticks(X,['silver','gold','platinum'])
plt.ylabel('Prices')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([10,20,30])
Y=np.array([80,60,100])
Y2025=np.array([50,140,80])
plt.barh(X+1.5,Y,height=3,color='r',label='2024')
plt.barh(X-1.5,Y2025,height=3,color='b',label='2025')
plt.title('Goods Prices')
plt.yticks(X,['silver','gold','platinum'])
plt.ylabel('Prices')

plt.legend()
plt.show()
```
